*** Side note ***
important statistical concepts

Central tendency - location of middle values
    - mean or average
    - median - middle value 
        - used when we have a highly squewd distribution
    - mode - most frequently occuring value
        - generally used for categorical 
Spread - how far from the middle values generally fall in a data distribution
    - variance - mean of the squares of the individual deviations
    - standard deviation - square root of the variance - has the same units as the original data. Generally used when there is a normal distribution
    - interquartile range - the quartile 3 value minus the qurtile 1 value. Considered the most robust measure of spread because it is the least affected by outliers in the data
        - quartile - 3 values that evenly divide the distribution into 4 parts. 25% of the data falls below the first value, 50% falls below the 2nd value, and 75% falls below the 3rd value
Skewness - Measure of symmetry in the data distribution. Measures how evenly the data falls around the middle value
Kurtosis - Measure of how much of the data falls on the tails/edges of the distribution. Can tell us how many outliers are in the data set. For normal distribution the value is 3

Multivariant analysis - looks at how 2 or more variables are related to eachother. 
create a heatmap of the correlation coefficient of the quantitative variables

```python
df.columns
df_corr = df[['popularity', 'duration_ms', 'danceability', 'energy', 'key', 'loudness',
       'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness',
       'valence', 'tempo', 'time_signature']].corr()
df_corr
```

```python
fig, ax = plt.subplots(figsize=(15, 15))
sns.heatmap(df_corr, annot=True, cmap='mako')

#darker cell means more negatively correlated
```
